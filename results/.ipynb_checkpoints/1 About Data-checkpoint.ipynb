## 流动性

参考这篇文章，里面提到了下面说到的两种流动性指标，写paper的时候，流动性的定义可以在这里比较简单的找到

* Lin H, Wang J, Wu C. Liquidity risk and expected corporate bond returns[J]. Journal of Financial Economics, 2011, 99(3): 628-650.

数据：标普S&P 1950-2018月度数据

模型：Amihud liquidity
$$L_t=\frac{V_t}{|lnp_t-lnp_(t-1) |}$$
用Amihud（2002）提出的流动性比率作为股票市场的流动性指标
其中V_t表示第t日的交易额
P_t为第t日的收盘价
则流动性指标L_t表示的是当天的交易额除上当天对数收益率的绝对值

操作：
取标普月度close那一列作为收盘价
交易额计算：均价*Volume（其中均价=（High+Low）/2）

```python
import math
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.decomposition import PCA
from sklearn import preprocessing
%matplotlib inline

def check_data(inoutdat, issave = False):
    sns.set_style("white")

    num_plots = inoutdat.shape[1]
    n = int(math.ceil(math.sqrt(num_plots)))

    fig = plt.figure(figsize=(16, 6))
    axes = [plt.subplot(n, n, i) for i in range(1, num_plots + 1)]

    i = 0
    for k, v in inoutdat.iteritems():
        ax = axes[i]
        sns.kdeplot(v, shade=True, ax=ax, legend=False)
        [label.set_visible(False) for label in ax.get_yticklabels()]
        ax.xaxis.set_ticks([v.min(), v.max()])
        ax.set_title(k)
        i += 1
    sns.despine(left=True, trim=True, fig=fig)
    plt.tight_layout()
    
    if issave:
        plt.savefig("check_data.png")

def correlation(inoutdat, issave = False):
    sns.set(style="white")
    
    # Compute the correlation matrix
    corr = inoutdat.corr()
    
    # Generate a mask for the upper triangle
    mask = np.zeros_like(corr, dtype=np.bool)
    mask[np.triu_indices_from(mask)] = True
    
    # Set up the matplotlib figure
    f, ax = plt.subplots(figsize=(7, 7))
    
    # Generate a custom diverging colormap
    cmap = sns.diverging_palette(220, 10, as_cmap=True)
    
    # Draw the heatmap with the mask and correct aspect ratio
    sns.heatmap(corr, mask=mask, cmap=cmap, vmax=.3, center=0,
                square=True, linewidths=.5, cbar_kws={"shrink": .5})
    if issave:
        plt.savefig("correlation.png")
```
